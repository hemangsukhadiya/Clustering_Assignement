# **Clustering -  Theoretical Questions :**

### 1. What is unsupervised learning in the context of machine learning?


- Unsupervised learning involves training models on unlabeled data.

- The goal is to find hidden patterns or groupings in the data.

- Examples include clustering and dimensionality reduction.

----


### 2. How does K-Means clustering algorithm work?

- K-Means partitions data into a predefined number of clusters (K).

- It assigns data points to the nearest cluster center and iteratively updates cluster centroids.

- The process repeats until convergence or no significant change in centroids.

---


### 3. Explain the concept of a dendrogram in hierarchical clustering.

- A dendrogram is a tree-like diagram used to visualize hierarchical clustering.

- It shows how clusters are merged or split at various levels of similarity.

- The y-axis represents the distance or dissimilarity between clusters.

---



### 4. What is the main difference between K-Means and Hierarchical Clustering?

- K-Means is a centroid-based clustering algorithm, whereas Hierarchical Clustering builds a tree structure (dendrogram) of clusters.

- K-Means requires the number of clusters (K) to be predefined, while Hierarchical Clustering does not.

- K-Means is faster for large datasets, whereas Hierarchical Clustering can capture more detailed relationships.

---



### 5. What are the advantages of DBSCAN over K-Means?

- DBSCAN can detect arbitrarily shaped clusters, unlike K-Means which assumes circular clusters.

- It can identify outliers as noise points, while K-Means assigns all points to clusters.

- DBSCAN doesn’t require specifying the number of clusters beforehand.

---



### 6. When would you use Silhouette Score in clustering?

- Silhouette Score measures how similar points are within their own cluster compared to other clusters.

- It is used to evaluate the quality of clustering results.

- A higher score indicates well-formed clusters, while a lower score suggests poor clustering.

---


### 7. What are the limitations of Hierarchical Clustering?

- Hierarchical Clustering is computationally expensive (O(n^2) complexity) for large datasets.

- It doesn’t scale well to high-dimensional data.

- Once a merge or split happens, it cannot be undone, leading to suboptimal clusterings in some cases.

---





### 8. Why is feature scaling important in clustering algorithms like K-Means?


- Feature scaling ensures that all features contribute equally to the distance calculations.

- Without scaling, features with larger ranges can dominate the clustering process.

- It helps improve the performance and accuracy of distance-based algorithms like K-Means.

---


### 9. How does DBSCAN identify noise points?


- DBSCAN identifies noise points as those that are not within the "epsilon" neighborhood of any other points.

- Noise points do not meet the density requirements of being part of a cluster.

- These points are labeled as outliers or noise.

---

### 10. Define inertia in the context of K-Means.


- Inertia measures the sum of squared distances between data points and their cluster centroids.

- A lower inertia indicates better clustering, but it can decrease as K increases, even if clusters are poor.

---




### 11. What is the elbow method in K-Means clustering?

- The elbow method involves plotting the inertia (within-cluster sum of squares) for different values of K.

- The "elbow" point, where the inertia starts to level off, indicates the optimal number of clusters.

---

### 12. Describe the concept of "density" in DBSCAN.

- In DBSCAN, density refers to the number of points within a given radius (epsilon).

- A high-density region forms a cluster, while regions with low density are considered noise.

- DBSCAN uses density to identify core points, border points, and noise.

---
### 13. Can hierarchical clustering be used on categorical data?

- Hierarchical Clustering can be adapted for categorical data using appropriate distance metrics (e.g., Hamming distance).
- However, it is more commonly applied to numerical data where distance measures like Euclidean are more effective.
- For categorical data, specialized algorithms like k-mode clustering may be more suitable.


---

### 14. What does a negative Silhouette Score indicate?

- A negative Silhouette Score indicates that the data points are likely assigned to the wrong clusters.
- The point is closer to a neighboring cluster than its own, suggesting poor clustering.
- This suggests that the clustering structure may need improvement.

---

### 15. Explain the term "linkage criteria" in hierarchical clustering.

- Linkage criteria define how the distance between clusters is calculated in hierarchical clustering.
- Common methods include single linkage (minimum distance), complete linkage (maximum distance), and average linkage (mean distance).
- The choice of linkage affects the shape and structure of the resulting clusters.


---


### 16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

- K-Means assumes that clusters are roughly spherical and of equal size.
- If clusters vary in size or density, K-Means may misassign points or fail to find optimal cluster centroids.
- It also struggles with non-convex clusters.


---

### 17. What are the core parameters in DBSCAN, and how do they influence clustering?

- The core parameters in DBSCAN are *epsilon* (maximum distance for points to be considered neighbors) and *minPts* (minimum number of points required to form a dense region).
- These parameters influence how DBSCAN detects clusters and noise points.
- Proper tuning is essential for accurate clustering results.

---

### 18. How does K-Means++ improve upon standard K-Means initialization?

- K-Means++ improves the initialization of centroids by selecting initial points that are farther apart, reducing the risk of poor initialization.
- It ensures better convergence by starting with more representative centroids.
- This method leads to better clustering results and faster convergence compared to random initialization.

---

### 19. What is agglomerative clustering?

- Agglomerative clustering is a type of hierarchical clustering where each data point starts as its own cluster.
- It repeatedly merges the closest pairs of clusters until a stopping criterion is reached.
- It’s a bottom-up approach, as opposed to divisive clustering, which is top-down.


---


### 20. What makes Silhouette Score a better metric than just inertia for model evaluation?

- Unlike inertia, which can decrease with more clusters (even if the clustering is poor), the Silhouette Score considers both cohesion and separation of clusters.
- A higher Silhouette Score reflects well-formed, distinct clusters.
- It helps evaluate the quality of the clusters, not just the distance between points and centroids.


----

```python
# 21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=300, centers=4, random_state=42)
kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='x')
plt.title("K-Means Clustering with 4 Centers")
plt.show()
```

```python
# 22. Load the Iris dataset and use Agglomerative Clustering to group the data into 3 clusters. Display the first 10 predicted labels.
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering

iris = load_iris()
X = iris.data
agg_clust = AgglomerativeClustering(n_clusters=3)
labels = agg_clust.fit_predict(X)

print("First 10 predicted labels:", labels[:10])
```

```python
# 23. Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot.
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

X, _ = make_moons(n_samples=300, noise=0.1, random_state=42)
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='Paired')
plt.title("DBSCAN on make_moons (Outliers Highlighted)")
plt.show()
```

```python
# 24. Load the Wine dataset and apply K-Means clustering after standardizing the features. Print the size of each cluster.
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
import numpy as np

wine = load_wine()
X = StandardScaler().fit_transform(wine.data)
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

unique, counts = np.unique(labels, return_counts=True)
print("Cluster sizes:", dict(zip(unique, counts)))
```

```python
# 28. Load the Digits dataset, reduce it to 2D using PCA, and visualize clusters from K-Means.
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA

digits = load_digits()
X = PCA(n_components=2).fit_transform(digits.data)
kmeans = KMeans(n_clusters=10, random_state=42)
labels = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='tab10', s=15)
plt.title("K-Means on Digits (PCA Reduced to 2D)")
plt.show()
```

```python
# 29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.
from sklearn.metrics import silhouette_score

X, _ = make_blobs(n_samples=300, centers=4, random_state=42)
scores = []
ks = range(2, 6)

for k in ks:
    labels = KMeans(n_clusters=k, random_state=42).fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)

plt.bar(ks, scores)
plt.xlabel("Number of clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for k = 2 to 5")
plt.show()
```

```python
# 30. Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage.
from scipy.cluster.hierarchy import dendrogram, linkage

Z = linkage(X, method='average')
plt.figure(figsize=(10, 5))
dendrogram(Z)
plt.title("Hierarchical Clustering Dendrogram (Average Linkage)")
plt.xlabel("Sample index")
plt.ylabel("Distance")
plt.show()
```

```python
# 31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.
from matplotlib.colors import ListedColormap
import numpy as np

X, _ = make_blobs(n_samples=300, centers=3, cluster_std=2.5, random_state=42)
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

h = 0.02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure()
plt.contourf(xx, yy, Z, cmap=ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF']), alpha=0.5)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title("K-Means with Overlapping Clusters and Decision Boundaries")
plt.show()
```

```python
# 34. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot.
X = load_breast_cancer().data
inertias = []

for k in range(2, 7):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X)
    inertias.append(km.inertia_)

plt.plot(range(2, 7), inertias, marker='o')
plt.xlabel("Number of clusters (K)")
plt.ylabel("Inertia")
plt.title("Inertia vs. K for Breast Cancer Dataset")
plt.show()
```
